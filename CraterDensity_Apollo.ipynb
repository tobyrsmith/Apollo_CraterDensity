```python
def log_10_product(x, pos):
    if (x < 1.0):
      return '%3.1f' % (x)
    else:
      return '%i' % (x)
```

```python
tfont = {
    'family' : 'DejaVu Serif',
    'weight' : 'bold',
    'size' : 14}

plt.rc('font', **tfont)
plt.rc("axes", linewidth=3.0)
plt.rc('axes', grid=False)

formatter = plt.FuncFormatter(log_10_product)
```

```python
def ringo(x,a,b,c):
    return a * (np.exp(b*x) - 1.0) + (c * x)

def george(x,c):
    return c*x

def john(x,a,b):
    return a * (np.exp(b*x) - 1.0)
```

```python
def PlotN(n, o, Fit = False, Save = False, Crust = False):
    
    A = 7.00875281e-15
    B = 9.58714635
    C = 2.57287271
    
    Z = np.linspace(0,5.0,1000)
    
    fig, ax = plt.subplots(1,1)
    fig.set_size_inches(10,7)

#    fig.tight_layout(pad=2.75)

    ax.grid(b=True, which='major')

    ax.set_yscale('log')
    ax.set_xlim(4.5,0)
    ax.set_ylim(0.1,1000)

    ax.yaxis.set_major_formatter(formatter)

    ax.set_xlabel('Age (billion years)')
    ax.set_ylabel('N(10)')
    
    ax.text(2.75,500,Name[n],fontdict=tfont,fontsize=18)
    
    if Fit:
        ax.plot(Z,ringo(Z,A,B,C),color='0.50',linestyle='-',linewidth=4)
    
    for i in o:
        ax.scatter(age[i],N[i],s=300,color=Color[i],alpha=0.5)
        
    if Crust:
        rect = pylab.Rectangle([4.05,90], 0.2, 900,facecolor='g', alpha = 0.5)
        ax.add_patch(rect)

    ax.hlines(N[n],6,age[n],linestyles='--',color=Color[n],linewidth=3)
    ax.vlines(age[n],0.01,N[n],linestyles='--',color=Color[n],linewidth=3)
    ax.scatter(age[n],N[n],s=300,color=Color[n],alpha=1.0);
    
    if Save:
        figname = Name[n] + ".png"
        fig.savefig(figname, dpi=300, bbox_inches='tight')
```

```python
#Apollo 11 - M.Tr.young

n = 5
o = [5]

PlotN(n,o,Save=True)
```
